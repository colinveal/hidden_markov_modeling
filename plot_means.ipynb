```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import kde
```

```python
sns.set(color_codes=True)
```

```python
ffile = "../"
```

```python
def collect_means(filename):
    
    with open(filename, 'r') as file:
        context = file.read()
        size = len(context)
        arraystr= context[1:size-1]
        arraystr = arraystr.split(',')
        means = [float(item) for item in arraystr]
        return means
    
```

```python
# load the non wga means

n_regions = 1
no_wga_wmeans = []
wga_wmeans = []
for region in range(n_regions):
    no_wga_wmeans.extend(collect_means(filename=ffile + "no_wga_windows_mean_" + str(region) + ".txt"))
    wga_wmeans.extend(collect_means(filename=ffile + "wga_windows_mean_" + str(region) + ".txt"))
```

```python
sns.distplot(no_wga_wmeans, bins=55, kde=False, rug=True)
plt.show()
```

```python
sns.distplot(wga_wmeans, bins=55, kde=False, rug=True)
plt.show()
```

```python
# visualize two 
nbins = 80
min_ = 0.0
max_ = 50.0

plt.figure( figsize=(10,10) )
plt.hist2d(no_wga_wmeans, wga_wmeans,
            bins=[nbins, nbins], cmap='Blues', density=False,
                 cmax=1000,
                 cmin=0,
                 alpha=0.99,
                 range=((min_, max_), (min_, max_)))
plt.title = "Hist 2D"
plt.show()
```
