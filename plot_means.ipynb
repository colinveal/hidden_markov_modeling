```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import kde
```

```python
sns.set(color_codes=True)
```

```python
ffile = "../"
```

```python
def collect_means(filename):
    
    
    with open(filename) as file:
        context = file.read()
        size = len(context)
        arraystr= context[1:size-1]
        arraystr = arraystr.split(',')
        means = [float(item) for item in arraystr]
        return means
    
```

```python
# load the non wga means

no_wga_wmeans_0 = collect_means(filename=ffile + "no_wga_windows_mean_0.txt")
no_wga_wmeans_1 = collect_means(filename=ffile + "no_wga_windows_mean_1.txt")
no_wga_wmeans_2 = collect_means(filename=ffile + "no_wga_windows_mean_2.txt")
no_wga_wmeans_3 = collect_means(filename=ffile + "no_wga_windows_mean_3.txt")

no_wga_wmeans = no_wga_wmeans_0 + no_wga_wmeans_1 + no_wga_wmeans_2 + no_wga_wmeans_3
```

```python
wga_wmeans_0 = collect_means(filename=ffile + "wga_windows_mean_0.txt")
wga_wmeans_1 = collect_means(filename=ffile + "wga_windows_mean_1.txt")
wga_wmeans_2 = collect_means(filename=ffile + "wga_windows_mean_2.txt")
wga_wmeans_3 = collect_means(filename=ffile + "wga_windows_mean_3.txt")

wga_wmeans = wga_wmeans_0 + wga_wmeans_1 + wga_wmeans_2 + wga_wmeans_3
```

```python
# visualize two 
nbins = 80
min_ = 0.0
max_ = 50.0

plt.figure( figsize=(10,10) )
plt.hist2d(no_wga_wmeans, wga_wmeans,
            bins=[nbins, nbins], cmap='Blues', density=False,
                 cmax=1000,
                 cmin=0,
                 alpha=0.99,
                 range=((min_, max_), (min_, max_)))
plt.title = "Hist 2D"
plt.show()
```
